# Standardization of Numerical Variables
‚óè	Objective: To scale numerical variables for uniformity, improving the dataset‚Äôs suitability for analytical models.
‚óè	Steps:
	Explain the concept of standardization (z-score normalization).
	Standardize the numerical columns using the formula: z=x-mu/sigma


Standardisation, also known as Z-score normalisation, is a crucial concept in data preprocessing, particularly in machine learning and statistics.
Technique used to scale the features of a dataset so that they have a mean of 0 and a standard deviation of 1. This transformation ensures that the
features contribute equally to the model and makes it easier to compare features with different scales.
This problem is solved by standardisation, which gives each feature a mean of 0 and a standard deviation of 1.

Formula :

z = (x - Œº) / œÉ

where:

z = standardized value, 
x = original value,
Œº = mean of the feature,
œÉ = standard deviation of the feature

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
numerical_cols = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
print(numerical_cols)
```

```python
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
def standardize(x):
    return (x - x.mean()) / x.std()

df[numerical_cols] = df[numerical_cols].apply(standardize)
```

```python
print(df[numerical_cols])
```

```python
df.hist(bins=10, figsize=(15, 10), color='blue', alpha=0.7)
plt.suptitle('Before Standardization', fontsize=16)
plt.show()
```

```python
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i+1)
    plt.hist(df[col], bins=50)
    plt.title(col)
plt.tight_layout()
plt.show()
```
